```python
# Una variable numérica i una categórica (Districte i Preu)
import seaborn as sns
```

```python
# Me generaba un error y en la columna de Preu he tenid que convertir los NAN en valor numérico.

df['Preu'] = pd.to_numeric(df['Preu'], errors='coerce')
```

```python
# Visualizo con un box plot ya que lo que me piden es la representación de una variable numérica con una categórica

sns.boxplot( x=df["Preu"], y=df["Nom_Districte"])
```

```python
# Guardo el gráfico como imagen.
plt.scatter(x=df['Trimestre'], y=df['Preu'], color='DarkSeaGreen')

plt.savefig('grafico3.png')
```

```python
# Crear un gráfico con tres variables (Barri o Districte, Trimestre i Preu). Voy a utilizar una visualización de barras apiladas. 
# Personalizo los colores a utilizar
colores = ['MidnightBlue','LightYellow','Teal','Salmon']

# Primero agrupo las dos variables categóricas con la numérica en una variable.
agrupado = df.groupby(['Nom_Districte', 'Trimestre'])['Preu'].sum().unstack()

# Luego, creo el gráfico de barras apiladas para mostrar el valor que tiene por districto en cada trimetre. 
ax = agrupado.plot(kind='bar', stacked=True, figsize=(10, 6), color=colores)
```

```python
# Guardo el gráfico en una imagen
gr5 = agrupado.plot(kind='bar', stacked=True, figsize=(10, 6), color=colores)
plt.savefig('grafico5')
```
